```python
import pandas as pd
import numpy as np

df = pd.read_csv("p3/stats.csv")

df = df.drop('Unnamed: 0', 1)

cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]

df
```

```python
df_train = df.loc[:228].copy()
df_test = df.loc[229:].copy()
```

```python
df_train = df_train[df_train.PTS != 0]
df_train = df_train[df_train.MIN != 0]
df_train = df_train[df_train.GP != 0]

df_test = df_test[df_test.PTS != 0]
df_test = df_test[df_test.MIN != 0]
df_test = df_test[df_test.GP != 0]


df_test
#df_train.loc[df['PTS'] == 0]
```

```python
df_train["log(PTS)"] = np.log(df_train["PTS"])
df_train
```

```python
import matplotlib.cm as cm

df_train.plot.scatter(x="MIN", y="GP", c="log(PTS)", cmap=cm.YlOrRd)
```

```python
X_train = df_train[["MIN", "GP"]]
y_train = df_train["log(PTS)"]

X_train_mean = X_train.mean()
X_train_sd = X_train.std()
X_train_st = (X_train - X_train_mean) / X_train_sd
```

```python
x_new = df_test.loc[230, ["MIN", "GP"]]

x_new_st = (x_new - X_train_mean) /X_train_sd
x_new_st
```

```python
dists = np.sqrt(((X_train_st - x_new_st) ** 2).sum(axis=1))
dists
```

```python
i_nearest = dists.sort_values().index[:5]
i_nearest
```
