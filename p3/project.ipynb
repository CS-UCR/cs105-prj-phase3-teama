```python
import pandas as pd
df = pd.read_csv("output.txt", sep=",")
df.head()
```

```python
df['full_name'] = df['first_name']+" "+df['last_name']
df = df.drop(["last_name","first_name"], axis=1)
df.head()
```

```python
df.to_csv(r'stats.csv') 
```

```python
dfplayer = pd.read_csv("players.csv")
dfplayer.head()
```

```python
dfplayer['full_name'] = dfplayer['first_name']+" "+dfplayer['last_name']
dfplayer = dfplayer.drop(["first_name","last_name"], axis=1)
dfplayer.head()
```

```python
dfplayer.to_csv(r'player.csv') 
```

```python
df_new = pd.merge(dfplayer, df, left_on='full_name', right_on='full_name' )
df_new.drop([r"Unnamed: 0"])

df_new
```

```python
df_new.to_csv(r'finaldata.csv') 
```
