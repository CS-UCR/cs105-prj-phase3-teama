```python
import pandas as pd
df = pd.read_csv("prediction_stats.csv")
df.fillna(0)
df.head()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
    
model = LinearRegression()
team = ["Milwaukee Bucks","Toronto Raptors","Boston Celtics","Miami Heat","Indiana Pacers","Philadelphia 76ers","Brooklyn Nets","Brooklyn Nets"]
x = len(team)
print("Round 1")
while len(team)>1:
    team1 = team[x-len(team)]
    team2 = team[x-len(team)+1]
    match = pd.DataFrame(columns=['PTS1','3PM1','REB1','STL1','PTS2','3PM2','REB2','STL2'], index=[0])
    random_scale = 4
    PTS1 = df.loc[df['name'] == team1, 'PTS'].mean()
    PM1 = df.loc[df['name'] == team1, '3PM'].mean()
    REB1 = df.loc[df['name'] == team1, 'REB'].mean()
    STL1= df.loc[df['name'] == team1, 'STL'].mean()
    AST1 = df.loc[df['name'] == team1, 'AST'].mean()

    PTS2 = df.loc[df['name'] == team2, 'PTS'].mean()
    PM2 = df.loc[df['name'] == team2, '3PM'].mean()
    REB2 = df.loc[df['name'] == team2, 'REB'].mean()
    STL2 = df.loc[df['name'] == team2, 'STL'].mean()
    AST2 = df.loc[df['name'] == team2, 'AST'].mean()
    
    match['PTS1'] = np.random.normal(PTS1, scale=random_scale)
    match['3PM1'] = np.random.normal(PM1, scale=random_scale)
    match['REB1'] = np.random.normal(REB1, scale=random_scale)
    match['STL1'] = np.random.normal(STL1, scale=random_scale)
    match['AST1'] = np.random.normal(AST1, scale=random_scale)

    match['PTS2'] = np.random.normal(PTS2, scale=random_scale)
    match['3PM2'] = np.random.normal(PM2, scale=random_scale)
    match['REB2'] = np.random.normal(REB2, scale=random_scale)
    match['STL2'] = np.random.normal(STL2, scale=random_scale)
    match['AST2'] = np.random.normal(AST2, scale=random_scale)

    
    match['PTS'] = match['PTS1'] - match['PTS2']
    match['3PM'] = match['3PM1'] - match['3PM2']
    match['REB'] = match['REB1'] - match['REB2']
    match['STL'] = match['STL1'] - match['STL2']
    match['AST'] = match['AST1'] - match['AST2']
    match = match[['PTS','3PM','REB','STL','AST']]
    
    X_train = match[["3PM", 'REB', 'STL', 'AST']]
    y_train = match["PTS"]
    model = LinearRegression()
    model.fit(X=X_train, y=y_train)
    match_array = match.values
    prediction = model.predict(X=X_train)
    if prediction > 0:
        print("winner is "+ team1)
        team.pop(x-len(team)+1)
    elif prediction < 0:
        print("winner is "+team2)
        team.pop(x-len(team))
    if len(team) == 4:
        print("Round 2")
        x =4
    if len(team) == 2:
        print("Conference finals")
        x =2
Ewinner = team[0]
```
