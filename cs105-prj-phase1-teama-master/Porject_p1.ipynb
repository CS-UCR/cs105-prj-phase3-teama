```python
import numpy as np
import pandas as pd
import requests
import json


team_list = []
i = 1
while i < 30:
    tid = str(i)
    JSONContent = requests.get("https://www.balldontlie.io/api/v1/teams/" + tid).json()
    if 'error' not in JSONContent:
        team_list.append([JSONContent['abbreviation'], JSONContent['city'], JSONContent['conference'],
                             JSONContent['division'], JSONContent['full_name']])
    i += 1
                         
dataset = pd.DataFrame(team_list)
dataset.columns = ['Abbreviation', 'City', 'Conference', 'Division', 'Name']
dataset.dropna(axis = 0, how = 'any', inplace = True)
dataset.index = pd.RangeIndex(len(dataset.index))
dataset.head(20)
```
