```python
import numpy as np
import pandas as pd
import requests
import json
import time


team_list = []
i = 1
while i < 494:
    tid = str(i)
    JSONContent = requests.get("https://www.balldontlie.io/api/v1/players/" + tid).json()
    if 'error' not in JSONContent:
        team_list.append([JSONContent['height_feet'], JSONContent['position'], JSONContent['first_name'],
                             JSONContent['last_name']])
    time.sleep(1)
    i += 1
          
dataset = pd.DataFrame(team_list)
dataset.columns = ['Abbreviation', 'City', 'Conference', 'Division']
dataset.dropna(axis = 0, how = 'any', inplace = True)
dataset.index = pd.RangeIndex(len(dataset.index))
dataset.head(-100)
```
